```python
import time
import random

def result(user, comp):
    if user == comp:
        res = "Its a TIE."
        return res
    
    if user == "s" and comp == "paper":
        res = "\nCOMPUTER WINS"
        return res
    
    if user == "p" and comp == "scissor":
        res = "\nCOMPUTER WINS"
        return res
    
    if user == "x" and comp == "stone":
        res = "\nCOMPUTER WINS"
        return res
    
    else:
        res = "\nWOOHOO...YOU WIN"
        return res   

def computer_turn():
    c_choice = ['scissor', 'paper', 'stone']
    selected_choice = random.choice(c_choice)
    return selected_choice

def game():
    
    while(True):
        choices = ['s', 'p', 'x']
        usr_choice= input("Your turn, please enter your choice : ").lower()

        if (usr_choice in choices):

                if usr_choice == "s":
                    print(f"You choose STONE.\n")
                    comp_choice = computer_turn()
                    print(f"Computer's turn...")
                    time.sleep(2)
                    print(f"Computer choose {comp_choice.upper()}.")
                    res = result(usr_choice, comp_choice)
                    print(res)

                if usr_choice == "p":
                    print(f"You choose PAPER.\n")
                    comp_choice = computer_turn()
                    print(f"Computer's turn...")
                    time.sleep(2)
                    print(f"Computer choose {comp_choice.upper()}.")
                    res = result(usr_choice, comp_choice)
                    print(res)

                if usr_choice == "x":
                    print(f"You choose SCISSOR.\n")
                    comp_choice = computer_turn()
                    print(f"Computer's turn...")
                    time.sleep(2)
                    print(f"Computer choose {comp_choice.upper()}.")
                    res = result(usr_choice, comp_choice)
                    print(res)

        else:
            print(f"\nInvalid input.\nPlease enter choice among s, p, x.")

        new_op = input("Do you want to play more (yes/no) : ")
        if new_op.lower() == "no":
                print("\nTHANKS FOR PLAYING THE GAME.")
                break
        else :
            continue
            
            

if __name__ == "__main__":

    print("============================================")
    print(f"WELCOME TO THE GAME STONE, PAPER, SCISSOR")

    rules = """Here are the rules:

1. Stone beats Scissor.
2. Scissor beats Paper.
3. Paper beats Stone.

You will be playing against the computer.
Enter your choice when prompted."""

    print("============================================")
    print(f"\n{rules}")

    print(f"""\nPress S for STONE.
Press P for PAPER.
Press X for SCISSOR.\n""")
    print("============================================")

    game()    
```
