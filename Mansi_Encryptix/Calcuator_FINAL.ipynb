```python
def is_float(value):
    try:
        float(value)
        return True
    
    except ValueError:
        return False
        

def calculator():
    print(f"Welcome to the calculator.\n") # \n for new line
    print(f"""Press 1 for addition.
Press 2 for substraction.
Press 3 for multiplication.
Press 4 for division.\n""")
    
    while(1):
        option = input("Input your choice : ")
        
        if option in ['1','2','3','4']:
            
            usr_input = input("Enter space seperated values : ")
            usr_input = usr_input.split(" ")
            
            numbers = []
            for num in usr_input:
                x = is_float(num)
                numbers.append(x)
                    
            if all(numbers):
                
                if option == "1":
                    print(f"\nOperation addition selected.")
                    sum = 0
                    for value in usr_input:
                        sum += float(value)
                    print(f"Sum : {sum}")
                    
                if option == "2":
                    print(f"\nOperation substraction selected.")
                    sub = float(usr_input[0])
                    for value in usr_input[1:]:  
                        sub -= float(value)
                    print(f"Subraction : {sub}")
                    
                if option == "3":
                    print(f"\nOperation multiplication selected.")
                    prod = 1
                    for value in usr_input:
                        prod *= float(value)
                    print(f"Multiplication : {prod}")
                    
                if option == "4":
                    print(f"\nOperation division selected.")
                    div = float(usr_input[0])
                    for value in usr_input[1:]:  
                        div /= float(value)
                    print(f"Division : {div}")
                    
            else:
                print("Please enter a valid entry.")
                
        else:
            print(f"Invalid option, please choose between 1 to 4.")
        
        new_op = input("Do you want to perform more operation (yes/no) : ")
        if new_op.lower() == "no":
            print("\nThank you for using Calculator.")
            break
        else:
            continue
            
calculator()
```

```python
import customtkinter as ctk
from tkinter import messagebox

def is_float(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def perform_operation():
    option = operation_var.get()
    usr_input = entry.get()
    usr_input = usr_input.split()
    
    if option in ['1', '2', '3', '4'] and usr_input:
        numbers = [is_float(num) for num in usr_input]
        
        if all(numbers):
            if option == "1":
                result = sum(float(value) for value in usr_input)
                operation = "addition"
                
            elif option == "2":
                result = float(usr_input[0])
                for value in usr_input[1:]:
                    result -= float(value)
                operation = "subtraction"
                
            elif option == "3":
                result = 1
                for value in usr_input:
                    result *= float(value)
                operation = "multiplication"
                
            elif option == "4":
                result = float(usr_input[0])
                for value in usr_input[1:]:
                    result /= float(value)
                operation = "division"
                
            messagebox.showinfo("Result", f"Operation {operation} selected.\nResult: {result}")
        else:
            messagebox.showerror("Error", "Please enter valid numbers.")
    else:
        messagebox.showerror("Error", "Invalid option or empty input. Please choose an operation and provide space-separated values.")

def reset_fields():
    operation_var.set(None)
    entry.delete(0, ctk.END)

app = ctk.CTk()
app.title("Calculator")
app.geometry("400x450")

ctk.CTkLabel(app, text="Select Operation:", font=("Helvetica", 16)).pack(pady=10)

operation_var = ctk.StringVar()

ctk.CTkRadioButton(app, text="Addition", variable=operation_var, value='1').pack(pady=5)
ctk.CTkRadioButton(app, text="Subtraction", variable=operation_var, value='2').pack(pady=5)
ctk.CTkRadioButton(app, text="Multiplication", variable=operation_var, value='3').pack(pady=5)
ctk.CTkRadioButton(app, text="Division", variable=operation_var, value='4').pack(pady=5)

ctk.CTkLabel(app, text="Enter space-separated values:", font=("Helvetica", 16)).pack(pady=10)
entry = ctk.CTkEntry(app, width=300)
entry.pack(pady=10)

ctk.CTkButton(app, text="Calculate", command=perform_operation, width=200).pack(pady=10)
ctk.CTkButton(app, text="Reset", command=reset_fields, width=200).pack(pady=10)
ctk.CTkButton(app, text="Exit", command=app.quit, width=200).pack(pady=10)

app.mainloop()
```
